```python
# 1: Prompt the user to enter three numbers
num1 = float(input("Please enter the first number: "))
num2 = float(input("Please enter the second number: "))
num3 = float(input("Please enter the third number: "))

# 2: Check if all three numbers are equal
if num1 == num2 == num3:
    if num1 > 0:
        print(f"All three numbers are equal and positive: {num1}")
    elif num1 < 0:
        print(f"All three numbers are equal and negative: {num1}")
    else:
        print(f"All three numbers are equal and zero: {num1}")
else:
    # 3: Find and print the maximum and minimum numbers
    max_num = max(num1, num2, num3)
    min_num = min(num1, num2, num3)
    print(f"The maximum number is: {max_num}")
    print(f"The minimum number is: {min_num}")
```

```python
# 1: Prompt the user to enter a word
word = input("Please enter a word: ")

# 2: Define vowels
vowels = 'aeiouAEIOU'
found = False

# 3: Check for vowels and print their positions
for index, char in enumerate(word):
    if char in vowels:
        print(f"Vowel '{char}' found at position {index}")
        found = True

if not found:
    print("No vowels found in the word.")
```

```python
#1 : Prompt the user to enter the total number of working days and absent days
total_days = int(input("Enter the total number of working days: "))
absent_days = int(input("Enter the total number of days absent: "))

#  2: Calculate the number of days attended
attended_days = total_days - absent_days

# 3: Calculate the attendance percentage
attendance_percentage = (attended_days / total_days) * 100

# 4: Print the attendance percentage and check if the student can sit for the exam
print(f"Attendance percentage: {attendance_percentage:.2f}%")
if attendance_percentage < 75:
    print("The student will not be able to sit in the exam.")
else:
    print("The student can sit in the exam.")
```

```python
# 1: Prompt the user to enter the marked price
marked_price = float(input("Please enter the marked price: "))

# 2: Determine the discount rate based on the marked price
if marked_price > 10000:
    discount_rate = 0.20
elif marked_price > 7000:
    discount_rate = 0.15
else:
    discount_rate = 0.10

# 3: Calculate the discount amount
discount_amount = marked_price * discount_rate

# 4: Calculate the final price after discount
final_price = marked_price - discount_amount

# 5: Print the final price
print(f"The final price after discount is: {final_price:.2f}")
```

```python
# 1: Prompt the user to enter a string
input_string = input("Please enter a string: ")

# 2: Count lower and upper case letters
lower_case_count = sum(1 for char in input_string if char.islower())
upper_case_count = sum(1 for char in input_string if char.isupper())

# 3: Convert string based on case counts
if lower_case_count < upper_case_count:
    converted_string = input_string.upper()
else:
    converted_string = input_string.lower()

# 4: Print the converted string
print(f"Converted string: {converted_string}")
```

```python
def is_palindrome(number):
    return str(number) == str(number)[::-1]

def is_armstrong(number):
    num_str = str(number)
    power = len(num_str)
    return sum(int(digit) ** power for digit in num_str) == number

# : Prompt the user to enter a number
number = int(input("Please enter a number: "))

# : Check if the number is palindrome or Armstrong
if is_palindrome(number):
    print(f"The number {number} is a palindrome.")
else:
    print(f"The number {number} is not a palindrome.")

if is_armstrong(number):
    print(f"The number {number} is an Armstrong number.")
else:
    print(f"The number {number} is not an Armstrong number.")
```

```python
# This requires a more complex implementation. Here's a basic version for numbers from 0 to 999:
def number_to_words(num):
    if num == 0:
        return "zero"
    
    units = ["", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
    teens = ["", "eleven", "twelve", "thirteen", "fourteen", "fifteen", "sixteen", "seventeen", "eighteen", "nineteen"]
    tens = ["", "ten", "twenty", "thirty", "forty", "fifty", "sixty", "seventy", "eighty", "ninety"]
    thousands = ["", "thousand"]
    
    def to_words(n):
        if n == 0:
            return ""
        elif n < 10:
            return units[n]
        elif n < 20:
            return teens[n-10]
        elif n < 100:
            return tens[n//10] + ("" if n % 10 == 0 else "-" + units[n % 10])
        elif n < 1000:
            return units[n//10] + " hundred" + ("" if n % 100 == 0 else " and " + to_words(n % 100))
    
    return to_words(num)

#  1: Prompt the user to enter a number
number = int(input("Please enter a number: "))
print(number_to_words(number))
```

```python
#Write a program to find the sum of the following series:
# 1 + 4 – 9 +16 – 25 + 36 ...........n terms
# 1: Prompt the user to enter the number of terms
n = int(input("Please enter the number of terms: "))

#2: Calculate the sum of the series
sum_series = sum((-1) ** (i % 2) * (i + 1) ** 2 for i in range(n))

#  3: Print the sum
print(f"The sum of the series is: {sum_series}")
```

```python
#  Count the Number of "car" Words in a String and Their Positions
#  1: Prompt the user to enter a string
input_string = input("Please enter a string: ")

#  2: Find and count the occurrences of "car"
word = "car"
count = input_string.count(word)
positions = [i for i in range(len(input_string)) if input_string.startswith(word, i)]

# 3: Print the count and positions
print(f"The word '{word}' appears {count} times.")
print(f"Positions: {positions}")
```
